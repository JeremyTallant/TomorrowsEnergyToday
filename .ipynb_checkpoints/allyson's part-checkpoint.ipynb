```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import gmaps
from scipy.stats import linregress

# Census & gmaps API Keys
from api import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Load the combined CSV file
combined_df = pd.read_csv("combined.csv")

# Display sample data
combined_df
```

```python
states_df= combined_df[["Name","solar_system_count_residential"]]
states_df.reset_index()
```

```python
northern_states_df= states_df.iloc[[33, 18, 23, 8, 41, 36, 39, 25, 4, 30, 12, 9]]
northern_states_df

```

```python
southern_states_df= states_df.iloc[[1, 15, 43, 22, 16, 37, 11, 6]]
southern_states_df
```

```python
northern_bar= northern_states_df.groupby(["Name"]).count()["solar_system_count_residential"]
panda_chart_1= northern_bar.plot(kind="bar", figsize=(6,6), color="blue")
plt.xlabel("States")
plt.ylabel("Solar System Count")
plt.tight_layout()
plt.show()
```

```python
plt.bar()
```

```python
southern_data =np.array(southern_states_df.groupby(["Name"]).count()["solar_system_count_residential"])
x_axis= np.arange(len(southern_data))

plt.bar(x_axis, southern_data, color="blue", alpha=0.5, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,["California", "Arizona", "New Mexico", "Texas", "Louisiana", "Mississippi", "Alabama", "Florida"],rotation='vertical')
plt.xlim(-0.75,len(x_axis)-0.25)
plt.ylim(0, max(x_axis)+0.5)
plt.xlabel("Name")
plt.ylabel("Solar Systems")
```

```python

# Display the map
gmaps.configure(api_key=gkey)
fig = gmaps.figure()
figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'}


# Display the map
locations= northern_states_df[["Name"]]
solar_n= northern_states_df["solar_system_count_residential"].astype(int)
solar_layer= gmaps.heatmap_layer(locations, weights=solar_n)
fig.add_layer(solar_layer)
fig
```
